```python
import pandas as pd

url="https://raw.githubusercontent.com/cheong-lieon/cheong-lieon-Maxis-Case-Study-Data-Insight/main/malaysia.csv"
data = pd.read_csv(url, header = None, skiprows=1)

data
```

```python
from scipy.spatial import distance
distance_airport_mal = distance.cdist(data.iloc[:, [4,5]], data.iloc[:, [4,5]], metric='euclidean')
distance_airport_mal
```

```python
distance_airport_mal_df = pd.DataFrame(distance_airport_mal)
distance_airport_mal_df
```

```python
distance_airport_mal_df.columns = data[1].values
distance_airport_mal_df.index = data[1].values
distance_airport_mal_df
```
